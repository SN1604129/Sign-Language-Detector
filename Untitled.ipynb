```python
import cv2
import time
labels = ['Hello', 'Yes', 'no', 'I', 'Love', 'You']

number_imgs = 45
for label in labels:
    !mkdir {'Tensorflow\workspace\images\collectedimages\\'+label}
    cap = cv2.VideoCapture(0)
    cap.set(3, 640)
    cap.set(4, 480)
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    for imgnum in range(number_imgs):
        IMAGES_PATH = 'D:\Tensorflow Object Detection\RealTimeObjectDetection\Tensorflow\workspace\images\collectedimages'+'\{}'.format(label)+'\{}.jpg'.format(str(imgnum))
        print('Image no {}'.format(imgnum))
        if cap.isOpened():
            ret, frame = cap.read()
            cv2.imwrite(IMAGES_PATH, frame)
            time.sleep(1)

        # load_image = cv2.imread('{}.jpg'.format(label))



    if cv2.waitKey(1) == 27:
        break

    cap.release()
    cv2.destroyAllWindows()
```
